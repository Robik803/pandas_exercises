### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
bars = chipo[["quantity","item_name"]].groupby("item_name").sum().sort_values("quantity", ascending=False).head(5)
plt.bar(bars.index,bars.quantity)
plt.xlabel("Items")
plt.xticks(rotation=90)
plt.ylabel("Quantity")
plt.title("Most ordered items")
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo.item_price = chipo.item_price.apply(lambda price: float(price.replace("$","")))
scat = chipo[["item_price","order_id","quantity"]].groupby("order_id").sum()
plt.figure()
plt.scatter(scat.item_price, scat.quantity)
plt.xlabel("Order Price")
plt.ylabel("Items Ordered")
plt.title("Number of items orderered per order price")
plt.show()
```
